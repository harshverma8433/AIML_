## ***Ordinal Encoding using Cyclic Line***


```python
df = pd.DataFrame({"Size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
```

```python
df.head(2)
```

```python
from sklearn.preprocessing import OrdinalEncoder

data = [["s", "m", "l", "xl"]] # The categories list needs to be a list of lists. Each inner list corresponds to the categories for a single feature

oe = OrdinalEncoder(categories=data) # Pass the correctly formatted categories list to the OrdinalEncoder
df["Size_en"] = oe.fit_transform(df[["Size"]])
```

```python
df
```

## ***Ordinal Encoding using Map function
***

```python
ord_data = {"s":4,"m":3,"l":7,"xl":8}
df["Size_en"] = df["Size"].map(ord_data)
df
```
