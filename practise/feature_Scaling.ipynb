```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv(r'/content/drive/MyDrive/datasets/loan_data_set.csv')
ds.head(3)
```

```python
ds.isnull().sum()
```

```python
ds.describe()
```

```python
sns.distplot(ds['ApplicantIncome'])
plt.show()
```

```python
sns.distplot(ds['CoapplicantIncome'])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
ss = sc.fit_transform(ds[["ApplicantIncome"]])
ds["ApplicantIncome_ss"] = pd.DataFrame(ss , columns=["x"])
ds.head(5)
```

```python
ds.describe()
```

```python
plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
sns.distplot(ds['ApplicantIncome'])
plt.subplot(1,2,2)
sns.distplot(ds['ApplicantIncome_ss'])
plt.show()
```

# min-max scaling(Normalization)

```python
ds.isnull().sum()
```

```python
ds.describe()
```

```python
sns.distplot(ds["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
ds["CoapplicantIncome_min"] = mms.fit_transform(ds[["CoapplicantIncome"]])
```

```python
ds.head(3)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.distplot(ds["CoapplicantIncome"])
plt.subplot(1,2,2)
sns.distplot(ds["CoapplicantIncome_min"])
plt.show()
```
