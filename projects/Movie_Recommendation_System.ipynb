```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("/content/drive/MyDrive/datasets/tmdb_5000_movies.csv")
movies.head(1)
```

```python
credits = pd.read_csv("/content/drive/MyDrive/datasets/tmdb_5000_credits.csv")
credits.head(1)
```

```python
movies = movies.merge(credits , on = "title")
movies.head(1)
```

```python
# genres
# id
# keywords
# title
# overview
# cast
# crew

movies = movies[["id" , "title" , "overview" , "genres" , "keywords" , "cast" , "crew" , "release_date"]]
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.head(1)
```

```python
movies["genres"][0]
```

```python
import ast  # convert string of list into list

```

```python
def convertintolist(object):
  l = []
  for i in ast.literal_eval(object):
    l.append(i.get("name"))
  return l
```

```python
movies['genres'] = movies['genres'].apply(convertintolist)
movies['keywords'] = movies['keywords'].apply(convertintolist)
movies.head(1)
```

```python
def fetch3Cast(object):
  l = []
  c = 0
  for i in ast.literal_eval(object):
    if c!=3:
      l.append(i.get("name"))
    else:
      break
  return l
```

```python
movies["cast"] = movies["cast"].apply(fetch3Cast)
movies.head(1)
```

```python
def fetchDirector(object):
  l = []
  for i in ast.literal_eval(object):
    if(i.get("job") == "Director"):
      l.append(i.get("name"))
      break
  return l
```

```python
movies["crew"] = movies["crew"].apply(fetchDirector)
movies.head(1)
```

```python
movies["overview"] = movies["overview"].apply(lambda x : x.split(" "))
movies["overview"][0]
```

```python
movies["genres"] = movies["genres"].apply(lambda x : [i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x : [i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x : [i.replace(" ","") for i in x])
movies["crew"] = movies["crew"].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
movies["tags"] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
```

```python
new_movies = movies[["id" , "title" , "tags" ,"release_date"]]
new_movies.head(1)
```

```python
new_movies["tags"] = new_movies["tags"].apply(lambda x : " ".join(x))
new_movies["tags"] = new_movies["tags"].apply(lambda x : x.lower())
```

```python
new_movies.head(1)
```

```python
# vectorization

# convert text to vector  it is called  textvectorization
# using bag of words  we conver it into vector
# so in this it map on the dimensional space so thee closest dimensions are the recommanded movies
# vector - 2d space - it is coordinates
# stop words is not considered
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000 , stop_words = "english")  # max_features tell how many words will take fot vectorization
cv.fit_transform(new_movies["tags"]).toarray() # # 4806 movies and transform in 5000
```

```python
for i in cv.get_feature_names_out():
  print(i)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem_text(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_movies["tags"] = new_movies["tags"].apply(stem_text)
```

```python
vectors = cv.fit_transform(new_movies["tags"]).toarray()
```

```python
for i in cv.get_feature_names_out():
  print(i)
```

```python
# find the cosine distance between the vectors . basically the vector is movies and each vector has 5000 numbers
# cosine - the angle between the two vector(movies)
# distance is inversly prop to similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity =  cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommand(movie):

  movie_index = new_movies[new_movies["title"] == movie].index[0]
  distances = similarity[movie_index]
  """
    similarity is assumed to be a 2D NumPy array or matrix where each element represents the similarity score between two movies.
similarity[movie_index] retrieves the row of similarity scores for the specified movie.
  """
  movies_list = list(enumerate(distances))
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movies_list:
    # print(i)
    print(new_movies.iloc[i[0]].title)
```

```python
recommand('Batman Begins')
```
